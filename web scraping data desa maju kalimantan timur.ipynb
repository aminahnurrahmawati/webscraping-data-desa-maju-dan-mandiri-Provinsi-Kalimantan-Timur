# Webscrapping data desa maju di provinsi kalimantan timur

Berikut merupakan pengolahan data mengenai desa maju di provinsi Kalimantan Timur yang diambil dari website DPMPD Kalimantan Timur. 
Desa maju sendiri merupakan Desa yang memiliki potensi sumber daya sosial, ekonomi, dan ekologi, serta kemampuan mengelolanya untuk peningkatan kesejahteraan masyarakat desa, kualitas hidup manusia, dan menanggulangi kemiskinan.

```python
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
```

```python
#webscrapping

link = "https://dpmpd.kaltimprov.go.id/desa/maju"
req = urllib.request.urlopen(link).read()
data = BeautifulSoup(req, 'html.parser')
table= data.find_all('div', {'class' : 'table-responsive'})[0] #line html tabel yang ingin diambil/discraping 
#(cara melihat : klik kanan mouse lalu klik inspect (Q))
```

```python
print(table)
```

```python
rows = table.find_all('tr')
```

```python
hasil= []

for row in rows:
    cols = row.find_all('td')
    cols= [ele.text.strip() for ele in cols]
    hasil.append(cols)
```

```python
df_hasil = pd.DataFrame(hasil)
df_hasil
```

```python
df_hasil.columns= ['No', 'Kabupaten', 'Kecamatan', 'Desa/Kelurahan', 'Strata']
df_hasil
```

```python
df_hasil.to_csv('data desa mandiri kaltim.csv', index=False) #menyimpan tabel hasil webscrapping ke file ekstensi csv
```

```python
df_hasil.isnull().sum() #checking missing value
```

```python
df_hasil= df_hasil.dropna() #eliminate missing value, because NaNs were in a row (No. 0)
```
