# Webscrapping data desa mandiri Provinsi Kalimantan Timur

Berikut merupakan pengolahan data mengenai desa mandiri di provinsi Kalimantan Timur yang diambil dari website DPMPD Kalimantan Timur.
Desa mandiri atau sangat maju merupakan desa maju yang memiliki kemampuan melaksanakan pembangunan desa untuk mnsejahterakan dan meningkatkan kualitas hidup masyarakatnya dengan ketahanan sosial, ketahanan ekonomi, dan ketahanan ekologgi yang berkelanjutan

```python
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
from IPython.display import HTML
```

```python
alamat = "https://dpmpd.kaltimprov.go.id/desa/mandiri"
req = urllib.request.urlopen(alamat).read()
data = BeautifulSoup(req, 'html.parser')
table= data.find_all('div', {'class' : 'table-responsive'})[0] #line html tabel yang ingin diambil/discraping 
#(cara melihat : klik kanan mouse lalu klik inspect (Q))
```

```python
print(table)
```

```python
rows = table.find_all('tr')
```

```python
#lakukan iterasi pada variabel row untuk memcah masing2 kolom menjadi sebuah list yg kemudian disimpan ke variabel hasil

hasil= []

for row in rows:
    cols = row.find_all('td')
    cols= [ele.text.strip() for ele in cols]
    hasil.append(cols)
```

```python
df_hasil = pd.DataFrame(hasil)
df_hasil
```

```python
df_hasil.columns= ['Kabupaten', 'Kecamatan', 'Desa/Kelurahan', 'Strata']
```

```python
df_hasil
```

```python
df_hasil.to_csv('data desa mandiri kaltim.csv', index=False) #menyimpan tabel ke file berekstensi csv
```
